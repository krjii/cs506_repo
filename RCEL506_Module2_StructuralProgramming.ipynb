```python
import ColabTurtle.Turtle as turtle

from google.colab import drive
drive.mount('/gdrive')

# Function to draw a regular polygon
def draw_polygon(sides, length):
    angle = 360 / sides
    for _ in range(sides):
        turtle.forward(length)
        turtle.right(angle)

# Initialize turtle
turtle.initializeTurtle()

# Draw a polygon
sides = 6  # Number of sides (example: hexagon)
length = 50  # Length of each side
draw_polygon(sides, length)

# Complete drawing
turtle.hideturtle()
```
